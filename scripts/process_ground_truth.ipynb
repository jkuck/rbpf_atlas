```python
objs = {}
prefix = 'training/label_02/'
for i in xrange(21):
    filename = "%04d.txt" % i
    with open(prefix+filename) as handle:
        for line in handle:
            data = line.split(" ")
            frame = int(data[0])
            track = int(data[1])
            obj_type = data[2]
            bbox = data[6:10]
            if obj_type != 'Car':
                continue
            center = [(float(bbox[0])+float(bbox[2])) / 2, (float(bbox[1])+float(bbox[3])) / 2]
            o = filename+str(track) # kind of lame hack but just appending id to *.txt filename as key
            if o not in objs.keys():
                objs[o] = [center]
            else:
                objs[o].append(center)
```

```python
json_objs = json.dumps(objs)
with open("groundtruth_tracks.json", "w") as saved_file:
    saved_file.write(json_objs)
```
